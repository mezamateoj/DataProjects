```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
books = pd.read_csv('./Datasets/clean_books.csv')
books.head()
```

```python
books.info()
```

```python
# change year to int
books['year'] = books['year'].astype(int)
```

```python
books.dtypes
```

```python
books['genre'].value_counts()
```

```python
books.describe()
```

```python
# plot histogram to look at the distribution of numerical data
# improve the plot by adding a title and labels and changing the color
sns.histplot(data=books, x='rating', color='red', binwidth=.1)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(data=books, x='year', color='red')
plt.title('Distribution of Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(data=books, x='year')
plt.show()
```

```python
sns.boxplot(data=books, x='year', y='genre')
plt.show()
```

## Lets explore another dataset

```python
unemployment = pd.read_csv('./Datasets/clean_unemployment.csv')
unemployment.head()
```

```python
unemployment.info()
```

```python
# select top 5 countries with the highest unemployment rate in 2021
top_5 = unemployment.sort_values(by='2021', ascending=False).head(5)
top_5
```

```python
# count missing values in each column
unemployment.isnull().sum()
```

```python
# select the columns that have missing values
unemployment[unemployment.isnull().any(axis=1)]
```

```python
unemployment.describe()
```

```python
unemployment['continent'].value_counts()
```

```python
# Create a histogram of 2021 unemployment; show a full percent in each bin
sns.histplot(data=unemployment, x='2021', binwidth=1)
plt.title('Distribution of Unemployment Rate in 2021')
plt.xlabel('Unemployment Rate')
plt.ylabel('Count')
plt.show()
```

```python
# Define a Series describing whether each continent is outside of South America
not_south_america = ~unemployment["continent"].isin(["South America"])

# Print unemployment without records related to countries in Oceania
unemployment[not_south_america]
```

```python
# Print the minimum and maximum unemployment rates during 2021
print(unemployment['2021'].min(), unemployment['2021'].max())

# Create a boxplot of 2021 unemployment rates, broken down by continent
sns.boxplot(data=unemployment, x='2021', y='continent')
plt.title('Unemployment Rate in 2021 by Continent')
plt.ylabel('Continent')
plt.xlabel('Unemployment Rate')
plt.show()
```
